```python
# Escribir un programa que pida al usuario su peso (en kg) y estatura (en metros), calcule el índice de masa corporal y lo almacene en una variable, y muestre por pantalla la frase Tu índice de masa corporal es <imc> donde <imc> es el índice de masa corporal calculado redondeado con dos decimales.

peso = float(input("Ingrese su peso en kg: "))               # Usamos en estos casos float porque con int los decimales se pierden.
altura = float(input("Ingrese su estatura en metros: "))

imc = ((peso)/(altura**2))
# imc = peso/(altura**2)
# imc = (peso)/(altura**2)

print(round(imc, 2))              #Para redondear un valor usamos el round en el print seguido del numero de decimales que queremos imprimir.

# round(5.1234, 2) ---> 5.12
# round(5.1274, 2)   ---> 5.13
# round(5.9999, 0)  ---> 5.0
```
