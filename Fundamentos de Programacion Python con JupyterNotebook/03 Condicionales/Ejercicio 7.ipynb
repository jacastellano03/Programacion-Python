```python
# Ejercicio 7
# Los tramos impositivos para la declaración de la renta en un determinado país son los siguientes: Escribir un programa que pregunte al usuario su renta anual y muestre por pantalla el tipo impositivo que le corresponde.

renta = float(input("Ingrese su renta anual: "))

if renta < 10000:
    print("Le corresponde un impuesto del 5%")

elif renta < 20000:
    print("Le corresponde un impuesto del 15%")

elif renta < 35000:
    print("Le corresponde un impuesto del 20%")

elif renta < 60000:
    print("Le corresponde un impuesto del 30%")

else:
    print("Le corresponde un impuesto del 45%")
    
```
