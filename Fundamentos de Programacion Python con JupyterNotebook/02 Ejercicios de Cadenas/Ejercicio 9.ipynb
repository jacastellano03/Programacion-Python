```python
# Ejercicio 9
# Escribir un programa que pregunte al usuario la fecha de su nacimiento en formato dd/mm/aaaa y muestra por pantalla, el día, el mes y el año. Adaptar el programa anterior para que también funcione cuando el día o el mes se introduzcan con un solo carácter.

fecha_de_nacimiento = input("Ingrese su fecha de nacimiento en formato dd/mm/aaaa:")
                    
posicion1 = fecha_de_nacimiento.find('/')
posicion2 = fecha_de_nacimiento.find('/',posicion1 +1)


dia = (fecha_de_nacimiento[:posicion1])
mes = (fecha_de_nacimiento[posicion1+1:posicion2])
año = (fecha_de_nacimiento[posicion2+1:])

print("Su fecha de nacimiento es el dia",dia,"del mes",mes,"del año",año)
print('Día', dia)
print('Mes', mes)
print('Año', año)

```
