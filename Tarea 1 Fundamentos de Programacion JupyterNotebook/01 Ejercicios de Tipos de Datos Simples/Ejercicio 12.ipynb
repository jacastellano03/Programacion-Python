```python
# Ejercicio 12
# Una panadería vende barras de pan a 3.49€ cada una. El pan que no es el día tiene un descuento del 60%. Escribir un programa que comience leyendo el número de barras vendidas que no son del día. Después el programa debe mostrar el precio habitual de una barra de pan, el descuento que se le hace por no ser fresca y el coste final total.

pan_del_dia_anterior_no_vendido = int(input("Ingrese el numero de barras de pan vendidas que son del día anterior: "))

print("Precio habitual por unidad barra de pan 3.49€")
costo_norm = (pan_del_dia_anterior_no_vendido * 3.49)
print("Precio final con descuento por unidad barra de pan 1,396€")
costo_con_desc = (3.49 * 0.60)
preciofinal = (3.49 - costo_con_desc)

costo_final_total = (pan_del_dia_anterior_no_vendido * preciofinal)

print("Precio total sin descuento: ",round(costo_norm,3))
print("Precio total con descuento: ",round(costo_final_total,3))
```
