```python
# Ejercicio 8
# Escribir un programa que pregunte por consola el precio de un producto en euros con dos decimales y muestre por pantalla el número de euros y el número de céntimos del precio introducido.

price_euro = input("Ingrese el precio del producto:")

entero = price_euro.find('.')

price_euro_int = (price_euro[:entero])
price_euro_flt = (price_euro[entero+1:])

print("El precio es de: "+price_euro_int+" euros con "+price_euro_flt+" centimos." )
```
